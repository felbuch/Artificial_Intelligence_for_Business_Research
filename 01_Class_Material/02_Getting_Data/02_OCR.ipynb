# OCR

OCR stands for __optical character recognition__.  
An OCR is a neural network that reads the text contained in an image.  

_Brainstorming session:_ why would this be useful for research?

__An OCR is a typical example of neural network you may want to use out-of-the-box, without any fine-tuning.__  (why?)

There are many, many OCRs out there. Some are paid, some are free. And some are better than others.

As of November 2022, here are some of the __top paid OCRs__:
 * Amazon Textract
 * Microsoft's cognitive service
 * Google cloud vision

And here are some of the __top free OCRs__:
* Easy OCR
* Pytesseract
* Keras-OCR
* Paddle OCR
* LayoutMv2 & LayoutMv3

_EasyOCR_ performs particularly well on structured pdf data, which is often the data we deal with in Business Research.  

So let's use this OCR to illustrate how OCRs work in general.

## Show image we'll be reading

```python
from PIL import Image
```

```python
image_path = "delete-me-5.jpg"

#Show
img = Image.open(image_path)
img
```

## Import and use OCR

```python
#!pip install easyocr
```

```python
#Import OCR
from easyocr import Reader
ocr = Reader(['en'])

#Use OCR
ocr_results = ocr.readtext(image_path)
```

```python
#Show OCR full results
ocr_results
```

```python
#Print text only
for element in ocr_results:
  print(element[-2])
```

```python
#Alternatively...
text = ' '.join([element[-2] for element in ocr_results])
text
```

## Save results

```python
with open('text_from_ocr.txt', 'w') as f:
    f.write(text)
```
